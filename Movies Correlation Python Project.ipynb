```python
# IMPORT LIBRARIES

import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the configuration of the plots

import seaborn as sns
import numpy as np
```

```python
# IMPORT DATA

df = pd.read_csv(r"C:\Users\PREYANSH\Desktop\DA PROJECT\Portfolio projects dataset\Movies python project\movies.csv",encoding='latin1')
```

```python
# GLANCE AT THE DATA

df.head()
```

```python
# ANY MISSING VALUES?

for col in df.columns:
    perc_mis = np.mean(df[col].isnull())
    print('{},{}'.format(col,perc_mis))
```

```python
# DATA TYPES

df.dtypes
```

```python
# CHANGE DATA TYPE OF BUDGET AND GROSS COLUMNS

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
# CORRECT THE YEAR COLUMN

df['corrected year'] = df['released'].astype(str).str[0:4]
df

df['year'] = df['corrected year']
df

del df['corrected year']
df
```

```python
# SORT GROSS FROM HIGHEST TO LOWEST

df = df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
# WHOLE VIEW OF THE DATASET

pd.set_option('display.max_rows', None)
df.sort_values(by = ['gross'], inplace=False, ascending=False)

df.head(15)
```

```python
# DROP DUPLICATES FROM COMPANY COLUMN

df['company'].sort_values(ascending = False)
df['company'].drop_duplicates().sort_values(ascending=False)
df.head(15)
```

```python
# SCATTER PLOT (BUDGET VS GROSS)

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')

plt.show()
```

```python
# PLOT REGRESSION BUDGET VS GROSS

sns.regplot(x= 'budget', y= 'gross', data=df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# LOOK AT THE CORRELATION

df.corr(method = 'pearson')
```

```python
# CREATE A HEAT MAP TO VISUALIZE THE CORRELATIONS

Correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(Correlation_matrix, annot = True)
```

```python
# CONVERT OBJECT TYPE TO NUMERIC

df_num = df

for col in df_num.columns:
    if (df_num[col].dtype == 'object'):
        df_num[col] = df_num[col].astype('category')
        df_num[col] = df_num[col].cat.codes

df_num.head(25)
```
